# Ensemble Methods
Ensemble methods in statistics and machine learning combine many learning algorithms to achieve greater predictive performance than any of the constituent learning algorithms alone. A machine learning ensemble, unlike a statistical ensemble in statistical mechanics, consists of only a specific limited collection of different models, but typically allows for considerably more flexible structure to exist within those alternatives.


Sequential ensemble techniques and parallel ensemble techniques are the two primary kinds of ensemble methods. Sequential ensemble approaches, such as Adaptive Boosting, create base learners in a sequential order (AdaBoost). The reliance between basic learners is promoted by the consecutive production of base learners. The model's performance is then improved by giving previously misrepresented learners larger weights.

In parallel ensemble approaches, such as random forest, basic learners are created in a parallel style. The simultaneous production of basis learners is used in parallel approaches to enhance base learner independence. The independence of base learners minimizes the mistake caused by the use of averages greatly.

In base learning, the majority of ensemble approaches use a single algorithm, resulting in homogeneity among all base learners. Homogenous base learners are those who are of the same kind and have similar characteristics. In other techniques, heterogeneous base learners are used, resulting in heterogeneous ensembles. Different sorts of learners make up heterogeneous base learners.

## Common types of ensembles
### Bayes optimal classifier
The Bayes optimal classifier is a classification technique. It is an ensemble of all the hypotheses in the hypothesis space. Each hypothesis is given a vote proportional to the likelihood that the training dataset would be sampled from a system if that hypothesis were true. The Bayes optimal classifier can be expressed with the following equation: $y={\underset {c_{j}\in C}{\mathrm {argmax} }}\sum _{h_{i}\in H}{P(c_{j}|h_{i})P(T|h_{i})P(h_{i})}$ <br>
The posterior is proportional to the likelihood times the prior:$ P(h_{i}|T)\propto P(T|h_{i})P(h_{i})$

### Boosting
Boosting entails forming an ensemble one step at a time by training each new model instance to emphasize the training examples that prior models misclassified. Boosting has been demonstrated to be more accurate than bagging in some circumstances, but it also has a higher risk of over-fitting the training data. Although some newer algorithms are said to generate greater results, Adaboost is by far the most prevalent implementation of boosting.

## 1. Bagging
Bagging method is a type of ensemble algorithm that creates several instances of a black-box estimator on random subsets of the original training set, then aggregates their unique forecasts to make a final prediction. These methods are used to lower the variance of a base estimator (for example, a decision tree) by introducing randomization into the construction mechanism and then constructing an ensemble from it. In many circumstances, bagging approaches provide a simple way to enhance with respect to a single model without having to change the underlying base algorithm. Bagging approaches, in contrast to boosting, operate best with powerful and complicated models (e.g., fully developed decision trees) because they provide a mechanism to reduce overfitting.

Bootstrapping and aggregation are two different methods of bagging. Bootstrapping is a sampling strategy that uses the replacement approach to obtain samples from the whole population (set). The sampling with replacement approach aids in the randomization of the selection process. To finish the method, the samples are run through the basic learning algorithm.

Aggregation is used in bagging to take into account all possible outcomes of the prediction and to randomize the result. Because all outcomes are not taken into account without aggregation, projections will be inaccurate. As a result, the aggregation is based on probability bootstrapping processes or the sum of all predictive model outputs.

* It improves the models accuracy
* It deals with higher dimensional data efficiently

## 3. AdaBoost
AdaBoost, short for Adaptive Boosting, is a boosting ensemble machine learning algorithm, and was one of the first successful boosting approaches. Although various methods can be utilized, AdaBoost combines the predictions from small one-level decision trees, known as decision stumps. The AdaBoost technique tries to employ multiple weak models and correct their forecasts by adding further weak models, hence decision stump procedures are used.

The approach was created for classification and includes merging the predictions from all of the ensemble's decision trees. For regression issues, a similar technique was devised, in which predictions are produced using the average of the decision trees. Each model's contribution to the ensemble prediction is weighted depending on its performance on the training dataset.

 boosted classifier is a classifier of the form

$F_{T}(x)=\sum _{t=1}^{T}f_{t}(x)\,\!$

the total training error $ E_{t}$ of the resulting $t$ t-stage boosted classifier is minimized.

$ E_{t}=\sum _{i}E[F_{t-1}(x_{i})+\alpha _{t}h(x_{i})]$
Here $ F_{t-1}(x)$ is the boosted classifier 

At each iteration of the training process, a weight $w_{i,t}$ is assigned to each sample in the training set equal to the current error $E(F_{t-1}(x_{i}))$ on that sample. 

## 4. Gradient Boosting
Gradient boosting is a machine learning technique that can be used for a variety of applications, including regression and classification. It returns a prediction model in the form of an ensemble of weak prediction models, most commonly decision trees.

**Algorithm**
Initialize model with a constant value:
$F_{0}(x)={\underset {\gamma }{\arg \min }}\sum _{i=1}^{n}L(y_{i},\gamma )$

Compute so-called pseudo-residuals:
$ r_{im}=-\left[{\frac {\partial L(y_{i},F(x_{i}))}{\partial F(x_{i})}}\right]_{F(x)=F_{m-1}(x)}\quad {\mbox{for }}i=1,\ldots ,n$

Fit a base learner and Compute multiplier $ \gamma _{m}$by solving the following one-dimensional optimization problem

Update the model:
$ F_{m}(x)=F_{m-1}(x)+\gamma _{m}h_{m}(x)$

**How Gradient Boosting Works**
1. Loss Function

It must be differentiable

2. Weak Learner

Decision trees are used as the weak learner in gradient boosting

3. Additive Model

A gradient descent procedure is used to minimize the loss when adding trees.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Set theme for plotting
sns.set_theme()

# Import the data


penguins = sns.load_dataset("penguins")

sns.jointplot(data=penguins, x="bill_depth_mm", y="bill_length_mm", hue="species")
sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g"
)
X = penguins[["bill_length_mm", "bill_depth_mm"]].to_numpy()
y = penguins["species"].map(make_labels).to_numpy()
```

**Plots shows basic classification of our dataset. The classification of species based on bill_depth_mm and bill_length_mm**

```python
from sklearn.preprocessing import StandardScaler
df =pd.read_csv("penguins_lter.csv")
df.fillna(df.mean())
df.dropna()
df.head()

culmen_columns = ["Culmen Length (mm)", "Culmen Depth (mm)"]
target_column = "Species"
         
X = df[culmen_columns].to_numpy()

def make_labels(y):
    if y == "Adelie Penguin (Pygoscelis adeliae)":
        return 0
    else:
        return 1
y= df[target_column].map(make_labels).to_numpy()  

#X = X.drop(X.columns[np.isnan(X).any()], axis=1)
#print(X,y)
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.4, 
                                                    random_state = 42)


from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier


stump_clf = DecisionTreeClassifier(max_depth=1, random_state=42)
stump_clf.fit(X_train, y_train)
stump_y_pred = stump_clf.predict(X_test)
print(f"Stump Classification Report")
print(classification_report(y_test, stump_y_pred), "\n")


tree_clf = DecisionTreeClassifier(max_depth=6, random_state=42)
tree_clf.fit(X_train, y_train)
tree_y_pred = tree_clf.predict(X_test)
print(f"Tree Classification Report")
print(classification_report(y_test, tree_y_pred), "\n")

bag_clf = BaggingClassifier(DecisionTreeClassifier(max_depth=1, random_state=42),
                            n_estimators = 500,
                            bootstrap = True,
                            n_jobs = -1)
bag_clf.fit(X_train, y_train)
bag_y_pred = bag_clf.predict(X_test)
print(f"Bagging Classification Report")
print(classification_report(y_test, bag_y_pred), "\n")
```

**A good F1 score approximates to 1. Here we have f1-scores are around 0.93-0.98 which means 3 models predict each observation correctly.**


```python
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize = (10, 8))
plot_decision_regions(X, y, clf = tree_clf)
plt.xlabel("feature x_0", fontsize = 15)
plt.ylabel("feature x_1", fontsize = 15)
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (10, 8))
plot_decision_regions(X, y, clf = bag_clf)
plt.xlabel("feature x_0", fontsize = 15)
plt.ylabel("feature x_1", fontsize = 15)
plt.grid()
plt.show()
```

```python
np.random.seed(0)
X = np.linspace(-1, 1, 100)
y = (X ** 4) + np.random.normal(-.09, .09, 100)
plt.figure(figsize=(8, 8))
plt.scatter(X, y, s=10)
plt.show()
```

**The range of data is [-1,1] with 100 scatters. The plot shows the distribution of the data**

```python
from sklearn.tree import DecisionTreeRegressor

X_new = X.reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X_new, 
                                                    y,
                                                    test_size = 0.4, 
                                                    random_state = 42)





tree_reg1 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg1.fit(X_train, y_train)

y2 = y_train - tree_reg1.predict(X_train)

tree_reg2 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg2.fit(X_train, y2)

y3 = y2 - tree_reg2.predict(X_train)

tree_reg3 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg3.fit(X_train, y3)
```

```python
# Optional hard coded final prediction
# y_pred = sum(tree.predict(X_new) for tree in (tree_reg1, tree_reg2, tree_reg3))
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,10))

ax1.scatter(X, y, s=10)
ax1.plot(X, tree_reg1.predict(X_new), color = "red")
ax1.set_title("h(x) = h_1(x)", fontsize = 18)

ax2.scatter(X, y, s=10)
ax2.plot(X, tree_reg1.predict(X_new)+tree_reg2.predict(X_new), color = "red")
ax2.set_title("h(x) = h_1(x) + h_2(x)", fontsize = 18)

ax3.scatter(X, y, s=10)
ax3.plot(X, tree_reg1.predict(X_new)+tree_reg2.predict(X_new)+tree_reg3.predict(X_new), color = "red")
ax3.set_title("h(x) = h_1(x) + h_2(x) + h_3(x)", fontsize = 18)
plt.show()
```
