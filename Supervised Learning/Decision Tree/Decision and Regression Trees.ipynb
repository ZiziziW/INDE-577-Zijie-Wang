## Decision Trees
### Introduction
The Decision Tree algorithm is part of the supervised learning algorithms family. The decision tree approach, unlike other supervised learning algorithms, may also be utilized to solve regression and classification issues. The purpose of using a Decision Tree is to create a training model that can use to predict the class or value of the target variable. It can learn simple decision rules from training data.

There are assumptions we make while using Decision tree. In the beginning, the whole training set is considered as the root. Feature values are preferred to be categorical. If the values are continuous, they must be discretized before the model can be built. Using a statistical approach, the order in which characteristics are placed as the tree's root or internal node is determined.

Terminology of Decision Trees:
* **Root Node**: The root node symbolizes the total population or sample, which is then separated into two or more homogeneous groups.
* **Splitting**: A process of dividing a node into two or more sub-nodes.
* **Decision Node**: A sub-node splits into further sub-nodes.
* **Branch / Sub-Tree**: A subsection of the entire tree is called branch or sub-tree.
* **Parent and Child Node**: A node is divided into sub-nodes is called a parent node of sub-nodes whereas sub-nodes are the child of a parent node.


This process is recursive and is repeated for every subtree rooted at the new node.

### Algorithm
Researchers suggested using some criteria like Entropy and Information gain. Criteria will calculate values for every attribute. The node splitting function has an impact on the decision tree's accuracy. 
#### **The Information gain function**

This is the information gain function formula. The formula states the information gain is a function of the entropy of a node of the decision tree minus the entropy of a candidate split at node t of a decision tree. Then the information gain of T for attribute a is the difference between the a priori Shannon entropy $\mathrm {H} (T)$ of the training set and the conditional entropy $\mathrm {H} {(T|a)}$. 

$IG(T,a)=\mathrm {H} (T)-\mathrm {H} (T|a)$ 

#### **Entropy** 
is a measure of the randomness based on the information being processed.  The higher the entropy means the harder to get any conclusions from that information.
 $$\mathrm {H} (X)=-\sum _{i=1}^{n}{\mathrm {P} (x_{i})\log \mathrm {P} (x_{i})}$$

 
 
 From the above graph, the entropy H(X) is zero when the probability is either 0 or 1. The Entropy is maximum when the probability is 0.5. This is the situation of maximum uncertainty since it is most difficult to predict the outcome of the next toss; the result of each toss of the coin delivers one full bit of information. This is because \begin{aligned}
 \mathrm {H} (X)&=-\sum _{i=1}^{n}{\mathrm {P} (x_{i})\log _{b}\mathrm {P} (x_{i})}\\&=-\sum _{i=1}^{2}{{\frac {1}{2}}\log _{2}{\frac {1}{2}}}\\&=-\sum _{i=1}^{2}{{\frac {1}{2}}\cdot (-1)}=1 \end{aligned}
 
#### **Gini impurity** 
is used by the CART (classification and regression tree) algorithm for classification trees.The Gini impurity can be computed by summing the probability $p_{i}$ of an item with label $i$ being chosen times the probability $\sum _{k\neq i}p_{k}=1-p_{i}$

#### **Gain ratio**
The gain ratio solves the problem of information gain by considering the number of branches that might result before splitting. It compensates for information gain by taking into consideration the intrinsic information of a split.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
df=pd.read_csv("diabetes_dataset.csv")
df.head()
```

```python
X = df[["BMI", "BloodPressure"]].to_numpy()
y = df["Outcome"].to_numpy()
plt.figure(figsize=(10,6))

plt.scatter(X[y==0][:,0],X[y==0][:,1],color="red",label="0")
plt.scatter(X[y==1][:,0],X[y==1][:,1],color="blue",label="1")
plt.xlabel('BMI')
plt.ylabel('BloodPressure')
plt.title('Diabetes')
plt.legend()
```

### Training, Validation, and Test Sets
The training set is applied to train, or fit, your model. The test set is needed for an unbiased evaluation of the final model. You shouldnt use it for fitting or validation. Here we use the function sklearn.model_selection.train_test_split() to split and to visualize the data.
https://realpython.com/train-test-split-python-data/

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.33, 
                                                    random_state=42)
c_train = ["red" if label == 0 else "blue" for label in y_train]
c_test = ["red" if label == 0 else "blue" for label in y_test]

plt.figure(figsize=(10, 8))
plt.scatter(X_train[:, 0], X_train[:, 1], c = c_train)
plt.xlabel("feature: x_0")
plt.ylabel("feature: x_1")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.33, 
                                                    random_state=42)
c_train = ["red" if label == 0 else "blue" for label in y_train]
c_test = ["red" if label == 0 else "blue" for label in y_test]

plt.figure(figsize=(10, 8))
plt.scatter(X_train[:, 0], X_train[:, 1], c = c_train)
plt.xlabel("feature: x_0")
plt.ylabel("feature: x_1")
plt.show()
```

### sklearn.tree.DecisionTreeClassifier

Parameters :

criterion : string, optional (default=gini)
The function to measure the quality of a split. Supported criteria are gini for the Gini impurity and entropy for the information gain.

max_depth : integer or None, optional (default=None)
The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.

min_samples_split : integer, optional (default=1)
The minimum number of samples required to split an internal node.

min_samples_leaf : integer, optional (default=1)
The minimum number of samples required to be at a leaf node.

min_density : float, optional (default=0.1)

max_features : int, string or None, optional (default=None)

compute_importances : boolean, optional (default=True)

```python
decision_tree = DecisionTreeClassifier(max_depth=3, random_state=42)
decision_tree.fit(X_train, y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(20, 8))
a = tree.plot_tree(decision_tree,
                   feature_names = ["feature: x_0", "feature: x_1"],
                   class_names =  ["red", "blue"],
                   rounded = True,
                   filled = True, 
                   fontsize=14)
plt.show()   
```

### For plotting the tree
sklearn.tree.plot_tree() function. The resulting tree has 13 nodes: 6 of them are split nodes and 6 are leaf nodes (or simply leaves), organized in 3 levels. 

### Decision boundaries
The following graph shows the decision boundaries from training data. The Decision Tree is by far the most sensitive, showing only extreme classification probabilities that are heavily influenced by single points. 

```python
# Set a plot_step
plot_step = 0.02


plt.figure(figsize=(10, 8))

# Plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

#plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

Z = decision_tree.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap='jet_r')

plt.scatter(X_train[:, 0], X_train[:, 1], c=c_train)
plt.xlabel("feature: x_0")
plt.ylabel("feature: x_1")
plt.title("Training Data")
```

### Confusion Matrix
Looking at the confusion matrix is a much better technique to assess a classifier's performance. The basic concept is to keep track of how many times examples of class A are categorised as class B. For example, you may check at the 5th row and 3rd column of the confusion matrix to see how many times the classifier confused images of 5s with images of 3s.

```python
# predicted values on the testing data
test_pred_decision_tree = decision_tree.predict(X_test)

# Import metrics from sklearn 
from sklearn import metrics

# Note: visualizing your tree above will be weird after running seaborn
import seaborn as sns

# The confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, test_pred_decision_tree)

# Convert confusion matrix into dataframe
matrix_df = pd.DataFrame(confusion_matrix)


plt.figure(figsize=(10, 8))
ax = plt.axes()
sns.set(font_scale=1.3)

sns.heatmap(matrix_df,
            annot = True, 
            fmt = "g",
            ax = ax, 
            cmap = "magma", 
            cbar = False)

ax.set_title("Confusion Matrix - Decision Tree")
ax.set_xlabel("Predicted Label", fontsize=15)
ax.set_xticklabels(["red", "blue"])
ax.set_ylabel("True Label", fontsize=15)
ax.set_yticklabels(["red", "blue"], rotation=0)
plt.show()
```

### The table of classification report 
The F1 is the weighted harmonic mean of precision and recall. The closer the value of the F1 score is to 1.0, the better the expected performance of the model is. Here we get 0.75 which means the accurarcy is high enough.

```python
from sklearn.metrics import classification_report

y_pred = decision_tree.predict(X_test)

#Checking performance our model with classification report.
print(classification_report(y_test, y_pred))
```

## Regression Trees
A regression tree is basically a decision tree that is used for the task of regression. It can be used to predict continuous valued outputs instead of discrete outputs. As another name for Decision Trees, which are used for regression. It can be beneficial in a variety of situations when the relationship between the variables is non-linear.
### Algorithm
**1.** Use recursive binary splitting to grow a large tree on the training data, stopping only when each terminal node has fewer
than some minimum number of observations.

**2.** Apply cost complexity pruning to the large tree to get a sequence of best subtrees 

**3.** Use K-fold cross-validation to choose 

**4.** Return the subtree with the value 

### Mean Square Error
We can't calculate the entropy and go through the same method because we're predicting continuous variables. Now we require a new metric. The entry-point of mean square error is a measure that shows us how far our predictions differ from the initial target.
$$\operatorname {MSE} ={\frac {1}{n}}\sum _{i=1}^{n}(Y_{i}-{\hat {Y_{i}}})^{2}$$
In matrix notation,
$$\operatorname {MSE} ={\frac {1}{n}}\sum _{i=1}^{n}(e_{i})^{2}={\frac {1}{n}}\mathbf {e} ^{\mathsf {T}}\mathbf {e}$$

###  Building in Scikit-learn
Here we use the same dataset(diabetes_dataset)to find the MSE value.




```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=3)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 12)

reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

reg1 = DecisionTreeRegressor(max_depth=1, random_state = 12)
reg1.fit(X_train, y_train)
y_pred = reg1.predict(X_test)
print(f"MSE reg1: {mean_squared_error(y_test, y_pred)} \n")


reg2 = DecisionTreeRegressor(max_depth=20, random_state = 12)
reg2.fit(X_train, y_train)
y_pred = reg2.predict(X_test)
print(f"MSE reg2: {mean_squared_error(y_test, y_pred)} \n")
```

### MSE value
The smaller the mean squared error, the best fit. In these two blocks, we can see that MSE values are around 0.2-0.3 which means the model fits well.
